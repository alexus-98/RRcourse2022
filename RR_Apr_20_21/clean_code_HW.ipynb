### Exercise 1 - FizzBuzz problem

```python
for i in range(1, 101):
    
    if i % 3 == 0 and i % 5 != 0:
        print('Fizz\n')
        
    elif i % 5 == 0 and i % 3 != 0:
        print('Buzz\n')
        
    elif i % 5 == 0 and i % 3 == 0:
        print('FizzBuzz\n')
        
    else:
        print(f'{i}\n')
```

### Exercise 2.1

#### Write a function which takes n (float or integer) as a parameter and returns the largest Fibonacci number smaller than n. The function should be documented, with type hints, and appropriate comments.

```python
def lastFromFibonacci(n):
    
    """
    Return the last number of Fibonacci Sequence that is smaller than n.
    Parameters: second_last and last stand for two last numbers in Fibonacci Sequence needed to compute another number.
    Initially they take values: 0 and 1 respectively.
    
    """

    second_last = 0
    last = 1
    
    # If n is bigger than the last tracked value, last two values are updated
    while n > last:
        
        
        # Sum of the last two numbers is stored
        temp = second_last + last
        
        # Values of parameters are updated
        second_last = last
        last = temp
    
    # If n is smaller than last tracked value of Fibonacci Sequence, the second last is returned
    else:
        return second_last
        
```

```python
lastFromFibonacci(100)
```

### Exercise 2.2

#### Write a function which reverses digits in an integer number (for example 7245 -> 5427). The function should be documented, with type hints, and appropriate comments.

```python
def numReverse(num):
    
    """
    This function reverses the order of digits in a given number.
    """
    
    # Cast to string
    num = str(num)
    
    # String is sliced and single digits are taken in reverse order
    num = num[::-1]
    
    # Cast back to integer
    num = int(num)
    
    return num
```

```python
numReverse(2345)
```
